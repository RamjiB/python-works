```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as py

x = np.array([np.arange(0,10,0.1)]).T #create a matrix with 100 values and tranposed it
m = 2 
c= 8
n = x.shape #assigning shape of x to n
y = m * x+c
ydis = y + np.random.random(n)*15 # disturbing the points so that it will show clearly
py.plot(x,ydis,"*")

#concatenate the two matrices so that we can multiple with m and c to find those values

a=np.hstack([x,np.ones(n)])

z = np.matmul(np.linalg.pinv(a),ydis) # finding m and c values
py.plot(x,np.matmul(a,z))

ydash = np.matmul(a,z)
e = ydis - ydash
np.dot(a.T ,e)
```
